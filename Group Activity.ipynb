```python
import pandas as pd
```

```python
df_titanic = pd.read_csv('Titanic dataset.csv')
df_titanic.head(5)
```

```python
#created a copy of the original dataset
df_titanic_copy = df_titanic
df_titanic_copy
```

```python
#remove duplicate values
df_titanic_copy.drop_duplicates(inplace=True)
```

```python
#drop name, home.dest, cabin, ticket columns
df_titanic_copy.drop(columns=['name', 'home.dest', 'cabin', 'ticket'], inplace=True, errors='ignore')

```

```python
#round up fare to 2 decimal places
df_titanic_copy['fare'] = df_titanic_copy['fare'].round(2)
```

```python
# convert categorical data into numerical. Male 0, Femal 1.
df_titanic_copy['sex'] = df_titanic_copy['sex'].map({'male': 0, 'female': 1})
```

```python
#drop rows with two or more null values
df_titanic_copy.dropna(thresh=df_titanic_copy.shape[1] - 2, inplace=True)

original_rows = df_titanic_copy.shape[0]

df_titanic_copy.dropna(thresh=df_titanic_copy.shape[1] - 2, inplace=True)


print(f"Number of rows dropped: {original_rows - (df_titanic_copy.dropna(thresh=df_titanic_copy.shape[1] - 2).shape[0])}")
```

```python
#fill null values in fares with mean value
df_titanic_copy['fare'].fillna(df_titanic_copy['fare'].mean(), inplace=True)
```

```python
#fill null age values with mean value
df_titanic_copy['age'].fillna(df_titanic_copy['age'].mean(), inplace=True)
```

```python
#check last 2 columns
# Display the first few rows to check the data
df_titanic_copy[['boat', 'body']].head()
```

```python
#drop rows where both body and boat are null
# Drop rows where both 'body' and 'boat' are NaN
df_titanic_copy.dropna(subset=['body', 'boat'], how='all', inplace=True)
```

```python
## Fill NaN values in 'body' where 'boat' has a value (not NaN)
df_titanic_copy['body'] = df_titanic_copy.apply(
    lambda row: 'Survived' if pd.notna(row['boat']) and pd.isna(row['body']) else row['body'], axis=1
)
```

```python
# Fill NaN in the 'boat' column with "No Boat" for those who did not survive
df_titanic_copy['boat'].fillna('No Boat', inplace=True)
```

```python
#save file 
df_titanic_copy.to_csv('cleaned_titanic.csv', index=False)
```
